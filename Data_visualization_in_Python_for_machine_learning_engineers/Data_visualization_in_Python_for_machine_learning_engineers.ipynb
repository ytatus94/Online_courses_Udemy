```python
import numpy as np
import matplotlib.pyplot as plt

def pdf(x, mu, sigma):
    a = 1. / (sigma * np.sqrt(2. * np.pi))
    b = -1. / (2. * sigma ** 2)
    return a * np.exp(b * (x - mu) ** 2)

x = np.linspace(-6, 6, 1000) # -6  6 ( -6  6 )  1000 

for i in range(5):
    samples = np.random.standard_normal(50)
    mu, sigma = np.mean(samples), np.std(samples)
    plt.plot(x, pdf(x, mu, sigma), color=".75") # color=0~1  0  1 
    
plt.plot(x, pdf(x, 0., 1.), color="k") # k 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pdf(x, mu, sigma):
    a = 1. / (sigma * np.sqrt(2. * np.pi))
    b = -1. / (2. * sigma**2)
    return a * np.exp(b * (x - mu)**2)

x = np.linspace(-6, 6, 1024)

plt.plot(x, pdf(x, 0., 1.),  color='k', linestyle='solid')
plt.plot(x, pdf(x, 0., .5),  color='k', linestyle='dashed')
plt.plot(x, pdf(x, 0., .25), color='k', linestyle='dashdot')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = 10**np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# lmplot()  figure level function figure
# grid = sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={'s': 100})
grid = sns.lmplot('x', 'y', data, height=7, truncate=True, scatter_kws={'s': 100}) # size  height 
grid.set_xticklabels(rotation=30) # tick  30 
plt.show()
```
